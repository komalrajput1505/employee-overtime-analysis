```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Employee': ['A','B','C','D','E','F','G','H'],
    'Department': ['IT','HR','IT','Finance','HR','IT','Finance','IT'],
    'Hours_Worked': [9,10,8,11,12,9,10,13],
    'Overtime': [1,1,0,1,1,0,1,1]
}

df = pd.DataFrame(data)
df
```

```python
overtime_count = df['Overtime'].value_counts()
overtime_count
```

```python
dept_overtime = df.groupby('Department')['Overtime'].sum()
dept_overtime
```

```python
sns.countplot(x='Department', hue='Overtime', data=df)
plt.title("Overtime by Department")
plt.show()
```

```python
plt.scatter(df['Hours_Worked'], df['Overtime'])
plt.xlabel("Hours Worked")
plt.ylabel("Overtime")
plt.title("Hours vs Overtime")
plt.show()
```
